Scenario: Employee Management System

You are tasked with developing an Employee Management System for a medium-sized company. The system should allow HR personnel to perform various tasks related to employee management efficiently.

Menu Options:

Add Employee: This option allows HR personnel to add a new employee to the system. They should be prompted to enter the employee's details such as name, age, position, and salary.

View Employee Details: This option enables HR personnel to view the details of a specific employee by entering their unique employee ID. Once entered, the system should display the employee's name, age, position, and salary.

Update Employee Details: HR personnel can use this option to update the details of an existing employee. They should be prompted to enter the employee's ID and then choose which information they want to update (name, age, position, or salary).

Delete Employee: This option allows HR personnel to remove an employee from the system. They should enter the employee's ID, and upon confirmation, the employee's details should be deleted from the system.

View Employee List: This option displays a list of all employees currently stored in the system, including their names and positions.

Exit: This option allows HR personnel to exit the Employee Management System.

Requirements:

Ensure proper validation of user inputs to prevent errors and ensure data integrity.
Implement error handling for cases such as invalid employee IDs or incorrect inputs.
Design the system to be user-friendly and intuitive, with clear prompts and messages guiding HR personnel through each task.
Note: You can implement this system in any programming language of your choice (e.g., Python, Java, etc.), utilizing appropriate data structures and control flow mechanisms.

```python
class EmployeeMS:
    
    def __init__(self):
        self.employee = {}
        
    def add(self,name,age,position,salary):
        emp_id = len(self.employee) + 1
        self.employee[emp_id] = {"name":name,"age":age,"position":position,"salary":salary}
        print(f"Employee has been added successfully with EMP ID : '{emp_id}'")
    
    def view(self,emp_id):
        if emp_id in self.employee:
            print(f"Emp Details: \nName: {self.employee[emp_id]['name']} | Age: {self.employee[emp_id]['age']} | Position: {self.employee[emp_id]['position']} | Salary: {self.employee[emp_id]['salary']}")
        
    def update(self,emp_id,ch,update):
        emp_info = self.employee.get(emp_id)
        if emp_info:
            if ch == 1:
                emp_info["name"] = update
                print(f"Emp ID : {emp_id} | Name updated successfully")
            elif ch == 2:
                emp_info["age"] = update
                print(f"Emp ID: {emp_id} | Age updated successfully")
            elif ch == 3:
                emp_info["position"] = update
                print(f"Emp ID: {emp_id} | position updated successfully")
            elif ch == 4:
                emp_info["salary"] = update
                print(f"Emp ID: {emp_id} | Salary updated successfully")
        else:
            print("Employee not found!")

            
    def delete(self,emp_id):
        del self.employee[emp_id]
        print(f"Emp ID: {emp_id} ,has been deleted.")
        
    def viewlist(self):
        for eid,info in self.employee.items():
            print(f"Emp ID: {eid}")
            print(f"Name: {info['name']} | Age: {info['age']} | Position: {info['position']} | Salary: {info['salary']}")

ob_ems = EmployeeMS()
print("-----::: Employee Management System :::-----")
while True:
    print("""
    1.Add Employee
    2.View Employee
    3.Update Employee
    4.View Employee List
    5.Delete Employee
    6.Exit
    """)
    ch = int(input("Enter your Choice:"))
    if ch == 1:
        name = input("Enter Name: ")
        age = int(input("Enter Age: "))
        position = input("Enter Position: ")
        salary = int(input("Enter salary: "))
        ob_ems.add(name,age,position,salary)
    
    elif ch == 2:
        emp_id = int(input("Enter Employee ID to view:"))
        ob_ems.view(emp_id)
        
    elif ch == 3:
        emp_id = int(input("Enter Employee ID to Update:"))
        print("""
        1.Update Name
        2.Update Age
        3.Update position
        4.Update Salary
        """)
        choice = int(input("Enter your Choice:"))
        if choice == 1:
            name = input("Enter name to update:")
            ob_ems.update(emp_id,choice,name)
        elif choice == 2:
            age = int(input("Enter age to update:"))
            ob_ems.update(emp_id,choice,age)
        elif choice == 3:
            position = input("Enter position to update:")
            ob_ems.update(emp_id,choice,position)
        elif choice == 4:
            salary = int(input("Enter salary to update:"))
            ob_ems.update(emp_id,choice,salary)
        else:
            print("Wrong Choice!")
    elif ch == 4:
        print("---:: Employee list ::---")
        ob_ems.viewlist()
        print("-------------------------")
    
    elif ch == 5:
        emp_id = int(input("Enter Employee ID to Delete:"))
        ob_ems.delete(emp_id)
    
    elif ch == 6:
        print("Exiting...")
        break
    
    else:
        print("Wrong choice! try again..")
```

Scenario: Library Management System

You've been tasked with developing a Library Management System to help librarians manage book borrowing and returning efficiently.

Menu Options:

Add Book: This option allows librarians to add a new book to the library's collection. They should be prompted to enter details such as the book's title, author, genre, and quantity.

Search Book: Librarians can search for a book in the library's collection by entering either the title or author's name. The system should display all matching books along with their details.

Borrow Book: Users can borrow a book from the library by entering the book's title and their name. If the book is available, the system should update its status to "borrowed" and record the borrower's name.

Return Book: Users can return a borrowed book by entering the book's title and their name. The system should update the book's status to "available" and remove the borrower's name.

View Borrowed Books: Librarians can view a list of all borrowed books along with the names of the borrowers.

View Book List: This option displays a list of all books currently available in the library, including their titles, authors, genres, and quantities.

Exit: This option allows users to exit the Library Management System.

Requirements:

Implement proper validation to ensure that users cannot borrow or return books that are not available or do not exist in the library's collection.
Design the system to handle multiple copies of the same book, updating the quantity available accordingly.
Provide clear and user-friendly prompts and messages to guide users through each task.
Ensure that the system maintains accurate records of borrowed books and their borrowers.
Note: You can implement this system in any programming language of your choice, using appropriate data structures and algorithms to manage the library's collection and user interactions.

```python
class LibraryMS:
    def __init__(self):
        self.books = []
        self.brlist = []
        
    def add(self,title,author,genre,quantity):
        for book in self.books:
            if book['title'] == title:
                book['quantity'] += quantity
                print("Book quantity updated")
                return
        
        self.books.append({'title': title, 'author': author, 'genre': genre, 'quantity': quantity})
        print("Book added")
    
    def search(self,name):
        found_books = []
        for book in self.books:
            if name.lower() in book['title'].lower() or name.lower() in book['author'].lower():
                found_books.append(book)
        if found_books:
            print("Found books:")
            for book in found_books:
                print(f"Title: {book['title']} | Author: {book['author']} | Genre: {book['genre']} | Quantity: {book['quantity']}")
        else:
            print("No books found with the given title or author.")
    
    def borrow(self,title,lname):
        for book in self.books:
            if book['title'] == title:
                if book['quantity'] > 0:
                    book['quantity'] -= 1
                    self.brlist.append({'title': title, 'lname': lname})
                    print("Book borrowed successfully!")
                    return
                else:
                    print("Sorry, this book is out of stock.")
                    return
        print("Book not found.")
        
    def returnb(self,title,lname):
        for book in self.brlist:
            if book['title'] == title and book['lname'] == lname:
                for b in self.books:
                    if b['title'] == title:
                        b['quantity'] += 1
                self.brlist.remove(book)
                print("Book returned successfully!")
                return
        print("Book not found in borrowed list.")
        
    def brview(self):
        for book in self.brlist:
            print(f"Book's Title: {book['title']} | Borrower's Name: {book['lname']}")
            
    def blview(self):
        for book in self.books:
            print(f"Book's Title: {book['title']} | Author: {book['author']} | Genre: {book['genre']} | Quantity: {book['quantity']}")

ob_lms = LibraryMS()
print("-----::: Library Management System :::-----")
while True:
    print("""
    1.Add Book
    2.Search Book
    3.Borrow Book
    4.Return Book
    5.View borrowed Book
    6.View Book list
    7.Exit
    """)
    ch = int(input("Enter your Choice:"))
    if ch == 1:
        title = input("Enter Title: ")
        author = input("Enter Author: ")
        genre = input("Enter Genre: ")
        quantity = int(input("Enter Quantity: "))
        ob_lms.add(title,author,genre,quantity)
    
    elif ch == 2:
        name = input("Enter Title or Author to search Book: ")
        ob_lms.search(name)
        
    elif ch == 3:
        title = input("Enter Book Title to Borrow: ")
        lname = input("Enter your Name: ")
        ob_lms.borrow(title,lname)
    
    elif ch == 4:
        title = input("Enter Book Title to Return: ")
        lname = input("Enter your Name: ")
        ob_lms.returnb(title,lname)
        
    elif ch == 5:
        print("---:: Borrowed Books ::---")
        ob_lms.brview()
        print("--------------------------")
    
    elif ch == 6:
        print("---:: Books List ::---")
        ob_lms.blview()
        print("----------------------")
    
    elif ch == 7:
        print("Exiting..")
        break
        
    else:
        print("Wrong choice! try again..")
```
