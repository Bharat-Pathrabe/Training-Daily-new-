Write a Python program to find the key of the maximum value in a dictionary.
Sample Output:<br>
Original dictionary elements:<br>
{'Theodore': 19, 'Roxanne': 22, 'Mathew': 21, 'Betty': 20}<br>
Finds the key of the maximum and minimum value of the said dictionary:<br>
('Roxanne', 'Theodore')

```python
def find(dict1):
    return max(dict1, key=dict1.get),min(dict1, key=dict1.get)

dict1 = {'Theodore': 19, 'Roxanne': 22, 'Mathew': 21, 'Betty': 20}
maxi,mini = find(dict1)
print(f"Maximum value Key: {maxi} | Minimum value Key: {mini}")
```

Write a Python program to transform a dictionary into a list of tuples.<br>
Sample Output:<br>
Original Dictionary:<br>
{'Red': 1, 'Green': 3, 'White': 5, 'Black': 2, 'Pink': 4}<br>
Convert the said dictionary to a list of tuples:<br>
[('Red', 1), ('Green', 3), ('White', 5), ('Black', 2), ('Pink', 4)]<br>

```python
def convert(odict):
    olist = []
    for key,value in odict.items():
        olist.append((key,value))
    return olist
    

odict = {'Red': 1, 'Green': 3, 'White': 5, 'Black': 2, 'Pink': 4}
clist = convert(odict)
print(f"Converted list: {clist}")
```

Write a Python program to combine two lists into a dictionary. The elements of the first one serve as keys and the elements of the second one serve as values. Each item in the first list must be unique and hashable.<br>
Sample Output:<br>
Original lists:<br>
['a', 'b', 'c', 'd', 'e', 'f']<br>
[1, 2, 3, 4, 5]<br>
Combine the values of the said two lists into a dictionary:<br>
{'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}

```python
def create(list1,list2):
    return dict(zip(list1,list2))

list1 = ['a', 'b', 'c', 'd', 'e', 'f']
list2 = [1, 2, 3, 4, 5]
dict1 = create(list1,list2) 
print(f"Combine : {dict1}")
```

Write a Python program to create a dictionary with the same keys as the given dictionary and values generated by running the given function for each value.<br>
Sample Output:<br>
Original dictionary elements:<br>
{'Theodore': {'user': 'Theodore', 'age': 45}, 'Roxanne': {'user': 'Roxanne', 'age': 15}, 'Mathew': {'user': 'Mathew', 'age': 21}}<br>
Dictionary with the same keys:<br>
{'Theodore': 45, 'Roxanne': 15, 'Mathew': 21}<br>

```python
def same_k(dict1,fn):
    dict2 = {}
    for ele,val in dict1.items():
        dict2[ele] = fn(val)
    return dict2

dict1 = {'Theodore': {'user': 'Theodore', 'age': 45}, 
         'Roxanne': {'user': 'Roxanne', 'age': 15}, 
         'Mathew': {'user': 'Mathew', 'age': 21}}
print(f"Same keys: {same_k(dict1, lambda u: u['age'])}")
```

Write a Python program to invert a dictionary with unique hashable values.<br>
Sample Output:<br>
{10: 'Theodore', 11: 'Mathew', 9: 'Roxanne'}

```python
def invert(dict1):
    return {value:key for key,value in dict1.items()}
dict1 = {'Theodore': 10, 'Mathew': 11, 'Roxanne': 9}
print(f"Inverted Dict: {invert(dict1)}")
```

Budgeting and expense tracking:<br>
It allows users to add expenses, view expenses, calculate total expenses, and analyze spending habits<br>

```python
class ExpenseTracking:
    def __init__(self):
        self.expense = []
    
    def add_expense(self, exp_cat, exp_amt):
        self.expense.append({"Category": exp_cat, "Amount": exp_amt})
        
    def view_expense(self):
        if not self.expense:
            print("No expenses")
        else:
            print(f"Category :\t Amount")
            for exp in self.expense:
                print(f"{exp['Category']} :\t {exp['Amount']}")
                
    def total(self):
        total = 0
        if not self.expense:
            print("No expenses")
        else:
            for exp in self.expense:
                total += exp['Amount']
        print(f"Total Expense: {total}")
        
    def analyze(self):
        if not self.expense:
            print("No expenses recorded yet.")
        else:
            categories = {}
            for exp in self.expense:
                category = exp["Category"]
                amount = exp["Amount"]
                if category in categories:
                    categories[category] += amount
                else:
                    categories[category] = amount

            print("Spending Habits Analysis:")
            for category, amount in categories.items():
                print(category, ": ", amount)

    

ob_et = ExpenseTracking()
print("-----::: Expense Tracker :::-----")
while True:
    print("""
    1.Add expense
    2.View expense
    3.Calculate total expense
    4.Analyze spending Habits
    5.Exit
    """)
    ch = int(input("Enter your choice: "))
    if ch == 1:
        exp_cat = input("Enter your Expense category: ")
        exp_amt = int(input("Enter your Expense Amount: "))
        ob_et.add_expense(exp_cat, exp_amt)
        print("Expense added successfully.")
        
    elif ch == 2:
        print("---: All Expenses :---")
        ob_et.view_expense()
        print("----------------------")
        
    elif ch == 3:
        ob_et.total()
    elif ch == 4:    
        ob_et.analyze()
    elif ch == 5:
        break
    else:
        print("Invalid choice! Choose again..")
```

simple contact list:<br>
1. Add Contact<br>
2. View Contacts<br>
3. Search Contact<br>
4. Remove Contact<br>
5. Exit

```python
class ContactList:
    def __init__(self):
        self.contacts = {}
        
    def add(self,name,num):
        self.contacts[name] = num
        
    def view(self):
        for key, value in self.contacts.items():
            print(f"Name: {key} | Number: {value}")

    def search(self,name):
        for key,value in self.contacts.items():
            if key == name:
                print(f"Name: {key} | Number: {value}")
                return None
        print("Contact not found")
    
    def remove(self,name):
        if name in self.contacts:
            del self.contacts[name]
        else:
            print("Not found")
    
ob_cl = ContactList() 
print("-----::: Contact List :::-----")
while True:
    print("""
    1.Add contact
    2.View contacts
    3.Search contact
    4.Remove contact
    5.Exit
    """)
    ch = int(input("Enter your choice: "))
    if ch == 1:
        name = input("Enter Contact name: ")
        num = int(input("Enter Contact Number: "))
        ob_cl.add(name,num)
        print("Contact added successfully.")
        
    elif ch == 2:
        print("---: All Contacts :---")
        ob_cl.view()
        print("----------------------")
        
    elif ch == 3:
        name = input("Enter contact name to search: ")
        ob_cl.search(name)
    elif ch == 4:    
        name = input("Enter contact name to remove: ")
        ob_cl.remove(name)
    elif ch == 5:
        break
    else:
        print("Invalid choice! Choose again..")
```

manage a to-do list:<br>
1. Add Task<br>
2. View Tasks<br>
3. Remove Task<br>
4. Mark Task as Done

```python
class ToDoList:
    def __init__(self):
        self.list = []
        
    def add(self,task):
        self.list.append(task)
        
    def view(self):
        for i,t in enumerate (self.list, start =1):
            print(f"{i}. {t}")
        
    def remove(self,task):
        if task >= 1 and task <= len(self.list):
            del self.list[task -1]
            print("Task removed")
        else:
            print("Task not found.")
        
    def mark(self,task):
        if task >= 1 and task <= len(self.list):
            self.list[task - 1] += " - Done"
            print("Marked done")
        else:
            print("Task not found")
    
ob_td = ToDoList() 
print("-----::: To-Do List :::-----")
while True:
    print("""
    1.Add Task
    2.View Tasks
    3.Remove Task 
    4.Mark task as done
    5.Exit
    """)
    ch = int(input("Enter your choice: "))
    if ch == 1:
        task = input("Enter task ")
        ob_td.add(task)
        print("Task added successfully.")
        
    elif ch == 2:
        print("---: All Tasks :---")
        ob_td.view()
        print("----------------------")
        
    elif ch == 3:
        task = int(input("Enter Task ID to remove: "))
        ob_td.remove(task)
    elif ch == 4:    
        task = int(input("Enter Task ID to mark done: "))
        ob_td.mark(task)
    elif ch == 5:
        break
    else:
        print("Invalid choice! Choose again..")
```

Basic library management system:<br>
1. Add Book:- title ,author ,copies<br>
2. Borrow Book<br>
3. Return Book<br>
4. View Books<br>
5. Exit

```python
class BasicLibraryManagementSystem:
    def __init__(self):
        self.lib = {}
        
    def add(self,title,author,copies):
        if title in self.lib:
            self.lib[title]["copies"] += copies
        else:
            self.lib[title] = {"author":author,"copies":copies}
        print("Book added successfully")
        
    def borrow(self,title):
        if title in self.lib:
            if self.lib[title]["copies"] > 0:
                self.lib[title]["copies"] -= 1
                print("Book borrowed Successfully")
            else:
                print("No book left to borrow")
        else:
            print("Book does not exist.")
        
    def returnb(self,title):
        if title in self.lib:
            self.lib[title]["copies"] += 1
            print("Book returned successfully")
        else:
            print("Book not found")
        
    def view(self):
        for title,info in self.lib.items():
            print(f"***\nTitle : {title} \nAuthor: {info['author']} \nCopies: {info['copies']}")

ob_bl = BasicLibraryManagementSystem() 
print("-----::: Basic library management system :::-----")
while True:
    print("""
    1.Add Book 
    2.Borrow Book
    3.Return Book 
    4.View Books
    5.Exit
    """)
    ch = int(input("Enter your choice: "))
    if ch == 1:
        title = input("Enter Book Title: ")
        author = input("Enter Book Author: ")
        copies = int(input("Enter number of copies: "))
        ob_bl.add(title,author,copies)
        
    elif ch == 2:
        title = input("Enter book title to borrow: ")
        ob_bl.borrow(title)
        
    elif ch == 3:
        title = input("Enter book title to return: ")
        ob_bl.returnb(title)
    
    elif ch == 4:    
        print("-----::: All Books :::-----")
        ob_bl.view()
        print("---------------------------")
    
    elif ch == 5:
        break
    
    else:
        print("Invalid choice! Choose again..")
```

Hospital Management System :<br>
    1. Add patient<br>
    2. View patients<br>
    3. Assign Doctor<br>
    4. View Doctors<br>
    5. Check patient status

```python
class Doctor:
    doctor_list = []

    def __init__(self, name, specialty):
        self.name = name
        self.specialty = specialty
        Doctor.doctor_list.append(self)

class Patient:
    patient_list = []

    def __init__(self, name, condition, assigned_doctor=None):
        self.name = name
        self.condition = condition
        self.assigned_doctor = assigned_doctor
        Patient.patient_list.append(self)

class HospitalManagementSystem:
    def __init__(self):
        pass 

    def add_patient(self,pat_name,condition):
        Patient(pat_name, condition)

    def view_patients(self):
        for pat in Patient.patient_list:
            doc_name = pat.assigned_doctor.name if pat.assigned_doctor else "Not Assigned"
            print(f"Name: {pat.name} | Condition: {pat.condition} | Doctor: {doc_name}")

    def assign_doctor(self,pat_name):
        for pat in Patient.patient_list:
            if pat.name == pat_name:
                doc_name = input("Enter doctor's name: ")
                specialty = input("Enter doctor's specialty: ")
                doctor = Doctor(doc_name, specialty)
                pat.assigned_doctor = doctor
                print("Doctor assigned successfully.")
                return
        print("Patient not found.")

    def check_patient(self,pat_name):
        for pat in Patient.patient_list:
            if pat.name == pat_name:
                doc_name = pat.assigned_doctor.name if pat.assigned_doctor else "Not Assigned"
                print(f"Name: {pat.name} | Condition: {pat.condition} | Doctor: {doc_name}")
                return
        print("Patient not found")

    def view_doctors(self):
        print("---:: Doctors List ::---")
        for doc in Doctor.doctor_list:
            print(f"Name: {doc.name} | Specialty: {doc.specialty}")
        print("-------------------------")


hms = HospitalManagementSystem()
print("-----::: Hospital Management System :::-----")
while True:
    print("""
    1. Add patient
    2. View patients
    3. Assign Doctor
    4. View Doctors
    5. Check patient status
    6. Exit
    """)
    ch = int(input("Enter your choice: "))
    if ch == 1:
        pat_name = input("Enter the patient name: ")
        condition = input("Enter the patient condition: ")
        hms.add_patient(pat_name,condition)
        print("Patient added successfully!")
        
    elif ch == 2:
        print("---:: Patients List ::---")
        hms.view_patients()
        print("-------------------------")
        
    elif ch == 3:
        pat_name = input("Enter the patient's name: ")
        hms.assign_doctor(pat_name)
        
    elif ch == 4:
        hms.view_doctors()
        
    elif ch == 5:
        pat_name = input("Enter patient's name to check: ")
        hms.check_patient(pat_name)
        
    elif ch == 6:
        print("Exiting Management System...")
        break
        
    else:
        print("Invalid choice.. Try again!")
```
