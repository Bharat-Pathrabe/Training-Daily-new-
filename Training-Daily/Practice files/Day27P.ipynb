```python
import tkinter as tk
from tkinter import colorchooser

def choose_color():
    color = colorchooser.askcolor(title="Select a Color")
    if color[1]:
        selected_color_label.config(text=f"Selected Color: {color[1]}")
        color_frame.config(bg=color[1])

parent = tk.Tk()
parent.title("Colorchooser Example")

# Create a button to open the color chooser dialog
choose_color_button = tk.Button(parent, text="Choose a Color", command=choose_color)
choose_color_button.pack(padx=20, pady=20)

# Create a label to display the selected color
selected_color_label = tk.Label(parent, text="Selected Color:", font=("Helvetica", 10))
selected_color_label.pack()

# Create a frame to display the selected color
color_frame = tk.Frame(parent, width=100, height=50)
color_frame.pack(padx=20, pady=20)

parent.mainloop()
```

```python
import re
def is_allowed_specific_char(string):
    charRe = re.compile(r'[^a-zA-Z0-9@]')
    string = charRe.search(string)
    return not bool(string)

print(is_allowed_specific_char("ABCDEFabcdef123450")) 
print(is_allowed_specific_char("Kerochan@555"))
```

Check if a string contains any special characters (!@#$%^&*()).

```python
import re

txt = input("Enter string: ")
x = re.search("[!@#$%^&*()]+",txt)
if x :
    print("Special characters exits")
else:
    print("No Special characters exits")
```

Validate if a string represents a valid email address.

```python
import re

txt = input("Enter string: ")
x = re.match("^[0-9a-zA-Z._+%-]+@[0-9a-zA-Z._-]+\.[a-zA-Z]{2,}$",txt)
if x :
    print("Valid email")
else:
    print("Invalid")
```

Determine if a string contains a valid URL.

```python
import re

def is_valid_url(input_string):
    pattern = r'https?://(?:www\.)?[a-zA-Z0-9-]+\.[a-zA-Z]{2}(?:/\S*)?'

    return re.match(pattern, input_string) is not None

# Test cases
urls = [
    "https://www.example.com",
    "http://example.com",
    "https://subdomain.example.com/path",
    "ftp://example.com",
    "invalidurl",
    "https://www.youtube.com/"
]

for url in urls:
    if is_valid_url(url):
        print("Valid URL:", url)
    else:
        print("Invalid URL:", url)
```

Python: Matches a string that has an a followed by zero or more b's

```python
import re
def text_match(text):
        patterns = '^a(b*)$'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')
            
txt = input("Enter: ")
print(text_match(txt))
```

Matches a string that has an 'a' followed by anything, ending in 'b'

```python
import re
def text_match(text):
        patterns = 'a.*?b$'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("aabbbbd"))
print(text_match("ababcbabab ab abbabb"))
print(text_match("accddbbjjjb"))
```
